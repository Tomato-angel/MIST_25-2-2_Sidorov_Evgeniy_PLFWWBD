## Задание №1 "Классический FizzBuzz"

### Стандартный подход
Используется конструкция if-elif-else

```python
n = int(input("Введите целое положительное число N: "))
for i in range(1, n + 1):
    # Стандартное решеине
    if i % 15 == 0:
        print("FizzBuzz", end = ' ')
    elif i % 3 == 0:
        print("Fizz", end = ' ')
    elif i % 5 == 0:
        print("Buzz", end = ' ')
    else:
        print(i, end = ' ')
```

### Альтернативный подход
Немноного видоизменённое решение


```python
n = int(input("Введите целое положительное число N: "))
for i in range(1, n + 1):
    # Альтернативный подход
    result = ""
    if i % 3 == 0: 
        result += "Fizz"
    if i % 5 == 0:
        result += "Buzz"
    if not result:
        result = f"{i}"
    print(f"{result}", end = ' ')
```

По похожему принципу решение можно упростить:

```python
for i in range(1, int(input("Введите целое положительное число N: ")) + 1):
    print("Fizz" * (i % 3 == 0) + "Buzz" * (i % 5 == 0) or i, end=' ')
```

# Задача 2: "Интерактивный менеджер списка покупок"

### Стандартный подход

Предполагаемое быстрое простое решение поставленной задачи

```python
shopping_list = []

while True:
    print("\n" + "="*30)
    print("Меню списка покупок".center(30))
    print("="*30)
    print("1. Добавить товар")
    print("2. Показать список")
    print("3. Выйти")
    print("="*30)
    
    command = input("Введите номер команды: ")
    
    if command == "1":
        item = input("Введите название товара: ")
        shopping_list.append(item)
        print(f"Товар '{item}' добавлен в список!")
        
    elif command == "2":
        if len(shopping_list) == 0:
            print("Список покупок пуст.")
        else:
            print("\nВаш список покупок:")
            for index, element in enumerate(shopping_list, 1):
                print(f"{index}. {element}")
                
    elif command == "3":
        print("Программа завершена. До свидания!")
        break
        
    else:
        print("Неизвестная команда, попробуйте снова.")
```

### Альтернативный подход

Более интересное решение с использованием отдельного класса для интерактивного списка покупок. 
Также тут более активно используются f-строки и различные другие "приколюхи" по выводу. 
Делал исключительно в рамках интереса посмотреть на то как реализуется подобное "ООП" в Python. 

**Краткое мнение/заметка (изложено в свободной форме):**
> Опыт и подходы конечно, крайне инетересные, однако требуют более детального размышления. (определять приватность и протектность подчёркиваниями в названии, серьёзно?, с непривычки такое очень сильно путает, да и читабельность падает. А что касается необходимости использования self., так и не разобрался, если без него, то поля станут статиками класса, как описал гугл, а если везде пихать этот self, то получится аналогия this. и привет C/C++, если есть способы улучшить читабельность в случаях подобного перегруза было бы инетересно ознакомиться, возможно я и вправду из-за нехватки опыта перебрал)

#### Описание класса для интерактивного списка

```python
class InteractiveShoppingList:
    '''Класс, ответственный за работу интерактивного списка покупок'''
    
    def __init__(self):
        self.__interactive_commands_list = []
        self.__shopping_list = []
        self.__is_executable = False
        
        self.__interactive_commands_list = [
            self.show_menu,
            self.enter_add_item,
            self.show_items,
            self.close
        ]
        
    def get_shopping_list(self):
        '''Возвращает список покукпок в виде массива'''
        return self.__shopping_list
    
    def show_menu(self):
        '''Выводит в консоль меню команд для работы со списком покупок'''
        menu = f'''
        \r{"="*50}
        \r{" Главное меню ".center(50, '-')}
        \r{"="*50}
        \r
        \r\033[1mПеречень команд ({len(self.__interactive_commands_list)})\033[0m:
        \r{'\n'.join(f"└─[ \033[1m{index}\033[0m ] : {element.__doc__};" for index, element in enumerate(self.__interactive_commands_list))}
        \r
        \r{"="*50}
        '''
        print(menu)
        
    def __show_undefined_command_message(self):
        '''Выводит сообщение о том, что команда не найдена и предлагает ввести команду повторно'''
        print("Неизвестная команда, повторите попытку ввода.")
        self.enter_command()

    def __check_list_closed(self):
        '''Проверяет закрыт (выключен) ли интерактивный список'''
        if not self.__is_executable:
            print("Список покупок закрыт")
            return True
        return False

    def __validate_command(self, string_index):
        '''Валидирует поступившее на вход текстовое представление команды (её индекса)'''
        index = -1
        if string_index.isdigit():
            index = int(string_index)
            if 0 <= index < len(self.__interactive_commands_list):
                return True, index
        return False, index
        
    def enter_command(self):
        '''Запрашивает ввод команды и исполняет её, предварительно валидируя'''
        if self.__check_list_closed(): 
            return 1
            
        command_string_index = input("Введите номер выбранной команды: ")
        is_valid, command_index = self.__validate_command(command_string_index)
        if is_valid:
            self.__interactive_commands_list[command_index]()
        else:
            self.__show_undefined_command_message()
        return 0
    
    def enter_add_item(self):
        '''Добавляет новый(-ые) товар(-ы) в список покупок'''
        if self.__check_list_closed(): 
            return 1
        items = input("Введите название(-ия) товара, которое следует поместить в список (при вводе нескольких названий, разделяйте их символом ','): ").split(',')
        for item in items:
            item = item.strip()
            while "  " in item:
                item = item.replace("  ", " ")  
            item = item.title()
            self.__shopping_list.append(item)
            print(f"Товар '{item}' добавлен в список покупок.")
        return 0
        
    def show_items(self):
        '''Показывает все товары в списке покупок'''
        if self.__check_list_closed(): 
            return 1
            
        if not self.__shopping_list:
            print("Список покупок пока пуст. Добавьте в него элементы, введя соответствующую команду.")
            return 0
            
        shopping_list_panel = f'''
        \r\033[1mНа текущий момент ваш список покупок состоит из:\033[0m
        \r{'\n'.join(f"\033[1m{index}\033[0m. {element}" for index, element in enumerate(self.__shopping_list))}
        '''
        print(shopping_list_panel)
        return 0
    
    def close(self):
        '''Завершает работу с списком покупок'''
        if self.__check_list_closed(): 
            return 1
        self.__is_executable = False
        print("\033[1;92m> Работа со списком покупок завершена.\033[0m")
        return 0
        
    def start(self):
        '''Запускает интерактивный режим работы со списком покупок'''
        try:
            if self.__is_executable: 
                print("Список покупок уже запущен, перед использованием необходимо закрыть работу с ним")
                return 1
                
            self.__is_executable = True
            print("\033[1;92m> Список покупок запущен!\033[0m")
            self.show_menu()
            while self.__is_executable:
                self.enter_command()
                
            return 0
        except Exception as e:
            print(f"\033[31mНепредвиденная ошибка в ходе работе класса интерактивного списка покупок {e}\n\033[0m")
```

#### Тестирование класса

```python
test = InteractiveShoppingList()
test.start()
```
