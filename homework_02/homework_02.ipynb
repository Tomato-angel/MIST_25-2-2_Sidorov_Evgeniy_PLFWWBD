## Задание №1 "Классический FizzBuzz"

### Стандартный подход
Используется конструкция if-elif-else

```python
n = int(input("Введите целое положительное число N: "))
for i in range(1, n + 1):
    # Стандартное решеине
    if i % 15 == 0:
        print("FizzBuzz", end = ' ')
    elif i % 3 == 0:
        print("Fizz", end = ' ')
    elif i % 5 == 0:
        print("Buzz", end = ' ')
    else:
        print(i, end = ' ')
print()
```

### Альтернативный подход
Немноного видоизменённое решение

```python
n = int(input("Введите целое положительное число N: "))
for i in range(1, n + 1):
    # Альтернативный подход
    result = ""
    if i % 3 == 0: 
        result += "Fizz"
    if i % 5 == 0:
        result += "Buzz"
    if not result:
        result = f"{i}"
    print(f"{result}", end = ' ')
print()
```
